```python
# input output seperation
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
# training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42) #object creation
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# normalization

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler() # object creation

# fit x_train
scaler.fit(x_train)

# converting input for normalization
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
# model creation
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)  #object creation and assigning value for k
knn.fit(x_train,y_train) # creating model using fit
```

```python
# to predict
y_pred=knn.predict(x_test)
y_pred #predicted outcome
```

```python
y_test  #actual outcome
```

```python
# performance evaluation
from sklearn.metrics import confusion_matrix,accuracy_score
cmd=confusion_matrix(y_test,y_pred) #object creation
cmd
```

```python
score=accuracy_score(y_test,y_pred)
score
```
