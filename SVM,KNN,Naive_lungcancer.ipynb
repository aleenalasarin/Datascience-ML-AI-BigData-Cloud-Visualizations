```python
df.drop(['Name'],axis=1,inplace=True)
```

```python
df.drop(['Surname'],axis=1,inplace=True)
df
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
# knn
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=7)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
score=accuracy_score(y_test,y_pred)
score
```

```python
# naive bayes
from sklearn.naive_bayes import GaussianNB
model1=GaussianNB()
model1.fit(x_train,y_train)
```

```python
y_pred1=model1.predict(x_test)
y_pred1
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm1=confusion_matrix(y_test,y_pred1)
cm1
```

```python
score1=accuracy_score(y_test,y_pred1)
score1
```

```python
# SVM
from sklearn.svm import SVC
model2=SVC()
model2.fit(x_train,y_train)
```

```python
y_pred2=model2.predict(x_test)
y_pred2
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm2=confusion_matrix(y_test,y_pred2)
cm2
```

```python
# using a for loop we can do it in single step
# first import
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score
# create object for each
knn=KNeighborsClassifier(n_neighbors=7)
bayes=GaussianNB()
svm=SVC()
# pass the objects name into a list
lst=[knn,bayes,svm]
```

```python
# modelcreation using for loop
for i in lst:
  print("Model name is",i) # model name is in i
  i.fit(x_train,y_train) #fit the model
  print("Prediction value....") # y value prediction
  y_pred=i.predict(x_test)
  print(y_pred)
  print("Confusion matrix is.....") #confusion matrix
  print(confusion_matrix(y_test,y_pred))
  print("Accuracy score is....") #accuracy score
  print(accuracy_score(y_test,y_pred))

```
