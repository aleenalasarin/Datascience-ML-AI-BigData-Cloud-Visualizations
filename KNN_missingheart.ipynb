```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/heart_missing.csv')
df
```

```python
x=df['cp'].mean()
x
y=df['fbs'].mean()
y
z=df['restecg'].mean()
z
b=df['oldpeak'].mean()
b
```

```python

df['cp'].fillna(x,inplace=True)
df['fbs'].fillna(y,inplace=True)
df['restecg'].fillna(z,inplace=True)
df['oldpeak'].fillna(b,inplace=True)
df.isna().sum()
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cmd=confusion_matrix(y_test,y_pred) #object creation
cmd
```

```python
score=accuracy_score(y_test,y_pred)
score
```
