```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/advertising.csv')
df
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
# seaborn : to grpahically plot each indenpendent variable, each dependent variablena eethallam reethiyil effect cheyyunnne graphically represent cheyyan
# regplot
import seaborn as sns
sns.regplot(x=df['TV'],y=y,color='g')
```

```python
sns.regplot(x=df['Radio'],y=y,color='c')
```

```python
sns.regplot(x=df['Newspaper'],y=y,color='m')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
# we cant draw the combined graph
print(model.predict([[200,40,50.6]]))
```

```python
df1=pd.DataFrame({'Actual value':y_test,'Predicted value':y_pred,'Difference':y_test-y_pred})
df1
```

```python
# slope
# here it prints 3 slopes (tv,radio,newspaper)
print("Slope is")
print(list(zip(x,model.coef_)))
```

```python
# constant  (only one constant)
print("Constant is",model.intercept_)
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
print("Error is",mean_absolute_error(y_test,y_pred))
```

```python
print("Error percentage",mean_absolute_percentage_error(y_test,y_pred))
```

```python
data=mean_squared_error(y_test,y_pred)
print(data)
```

```python
print("R2_score is",r2_score(y_test,y_pred))
```

```python
print(np.sqrt(data))
```
