```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['cap-shape']=lab.fit_transform(df['cap-shape'])
df['cap-surface']=lab.fit_transform(df['cap-surface'])
df['cap-color']=lab.fit_transform(df['cap-color'])
df['bruises']=lab.fit_transform(df['bruises'])
df['odor']=lab.fit_transform(df['odor'])
df['gill-attachment']=lab.fit_transform(df['gill-attachment'])
df['gill-spacing']=lab.fit_transform(df['gill-spacing'])
df['gill-size']=lab.fit_transform(df['gill-size'])
df['gill-color']=lab.fit_transform(df['gill-color'])
df['stalk-shape']=lab.fit_transform(df['stalk-shape'])
df['stalk-root']=lab.fit_transform(df['stalk-root'])
df['stalk-surface-above-ring']=lab.fit_transform(df['stalk-surface-above-ring'])
df['stalk-surface-below-ring']=lab.fit_transform(df['stalk-surface-below-ring'])
df['stalk-color-above-ring']=lab.fit_transform(df['stalk-color-above-ring'])
df['stalk-color-below-ring']=lab.fit_transform(df['stalk-color-below-ring'])
df['veil-type']=lab.fit_transform(df['veil-type'])
df['veil-color']=lab.fit_transform(df['veil-color'])
df['ring-number']=lab.fit_transform(df['ring-number'])
df['ring-type']=lab.fit_transform(df['ring-type'])
df['spore-print-color']=lab.fit_transform(df['spore-print-color'])
df['population']=lab.fit_transform(df['population'])
df['habitat']=lab.fit_transform(df['habitat'])

```

```python
df.dtypes
```

```python
x=df.iloc[:,1:].values
x
```

```python
y=df.iloc[:,0].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score,ConfusionMatrixDisplay
knn=KNeighborsClassifier(n_neighbors=7)
naive=GaussianNB()
svm=SVC()
lst=[knn,naive,svm]
```

```python
for i in lst:
  print("Model name is",i)
  i.fit(x_train,y_train)
  print("Prediction value.....")
  y_pred=i.predict(x_test)
  print(y_pred)
  print("Confusion matrix is....")
  cmd=confusion_matrix(y_test,y_pred)
  label=['p','e']
  cm=ConfusionMatrixDisplay(cmd,display_labels=label)
  print(cmd)
  print("Accuracy score is")
  print(accuracy_score(y_test,y_pred))
  print("confusion matrix display",i)
  cm.plot()
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_test,y_pred)
print(report)
```
