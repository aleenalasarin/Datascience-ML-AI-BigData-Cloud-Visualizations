```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/LoanApprovalPrediction.csv')
df
```

```python
df.drop(['Loan_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
x=df['Gender'].mode()[0]
x
y=df['Married'].mode()[0]
y
z=df['Dependents'].mode()[0]
z
a=df['Self_Employed'].mode()[0]
a
b=df['LoanAmount'].mean()
b
c=df['Loan_Amount_Term'].mean()
c
d=df['Credit_History'].mean()
d
```

```python
df['Gender'].fillna(x,inplace=True)
df['Married'].fillna(y,inplace=True)
df['Dependents'].fillna(z,inplace=True)
df['Self_Employed'].fillna(a,inplace=True)
df['LoanAmount'].fillna(b,inplace=True)
df['Loan_Amount_Term'].fillna(c,inplace=True)
df['Credit_History'].fillna(d,inplace=True)
df.isna().sum()
```

```python
df.dtypes
```

```python
df['Gender'].value_counts()
```

```python
df['Married'].value_counts()
```

```python
# no need of array when we do onehot encoding
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
# onehot encoding
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
col_trans=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['Gender','Married','Dependents','Education','Self_Employed','Property_Area']),remainder='passthrough')
x=col_trans.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score
knn=KNeighborsClassifier(n_neighbors=7)
naive=MultinomialNB()
svm=SVC()
lst=[knn,naive,svm]
```

```python
for i in lst:
  print("Model name is",i)
  i.fit(x_train,y_train)
  print("Prediction value is........ ")
  y_pred=i.predict(x_test)
  print(y_pred)
  print("Confusion matrix....")
  print(confusion_matrix(y_test,y_pred))
  print("Accuracy score is.....")
  print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_test,y_pred)
print(report)
```
