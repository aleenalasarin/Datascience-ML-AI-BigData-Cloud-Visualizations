```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/Salary_Data.csv')
df
```

```python
# seperate x and y
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
df.dtypes
```

```python
plt.scatter(x,y)
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.title('Year-Salary Graph')
```

```python
# training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# there is no need for normalization in regression model because when we do normalization it converts it into a smaller value
# model creation
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred  #predicted value
```

```python
y_test  #actual value
```

```python
# based on training datas we create a model(x_train,y_train)
plt.scatter(x_train,y_train)
plt.plot(x_test,y_pred,color='red')
```

```python
print(model.predict([[8.2]]))
```

```python
# y=mx+c  m=slope,c=constant,x=independent variable
# to print its slope and constant

print("slope is",model.coef_)  #model name.coefficient
```

```python
# to print constant

print("constant is",model.intercept_)
```

```python
# difference is
# actual value-predicted value
# create a dataframe
df1=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'Difference':y_test-y_pred})
df1
```

```python
# performance evaluation
# we should decrease the error vale,but r2_score has to be increased

from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error
print("error is",mean_absolute_error(y_test,y_pred))
print("percentage error is",mean_absolute_percentage_error(y_test,y_pred))
# accuracy about 94
```

```python
# r2_score
# we have to generate a higher value
from sklearn.metrics import r2_score
print("R2_score is",r2_score(y_test,y_pred))
```

```python
# MSE
from sklearn.metrics import mean_squared_error
data=mean_squared_error(y_test,y_pred)
print(data)
```

```python
# RMSE
# sqrt()
print(np.sqrt(data))
```
