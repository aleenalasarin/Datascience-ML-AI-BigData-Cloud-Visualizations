```python
x=df['Gender'].mode()[0]
x
df['Gender'].fillna(x,inplace=True)
df.isna().sum()
```

```python
x=df['Income'].mode()[0]
x
df['Income'].fillna(x,inplace=True)
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Gender']=lab.fit_transform(df['Gender'])
df
```

```python
df['Education']=lab.fit_transform(df['Education'])
df
```

```python
df['Marital Status']=lab.fit_transform(df['Marital Status'])
df
```

```python
df['Home Ownership']=lab.fit_transform(df['Home Ownership'])
df
```

```python
# input output seperation
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
# training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# normalization

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
# model creation
from sklearn.naive_bayes import BernoulliNB
model=BernoulliNB()
model.fit(x_train,y_train)
```

```python
# prediction
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
# performance_evaluation
from sklearn.metrics import confusion_matrix,accuracy_score,ConfusionMatrixDisplay
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
score=accuracy_score(y_test,y_pred)
score
```

```python
label=['High','Average','Low']
cmd=ConfusionMatrixDisplay(cm,display_labels=label)
cmd.plot()
```
