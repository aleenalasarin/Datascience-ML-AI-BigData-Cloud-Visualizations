```python
# Encoding
# get_dummies
# get_dummies is in pandas(encoding technique in pandas not in machine learning)
df5=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
df5
```

```python
# joining 2 dataframes
# df5 has only the values which we have encoded(18 columns) so we have to join both dataframes
# concatenation
dfe=pd.concat([df,df5],axis =1)
dfe
```

```python
# now dataframe name is dfe
# we have to drop unnamed,name,new_price
# drop fuel_type,transmission,owner_type,Location(beacause we encoded and we have the result data's)
dfe.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price'],axis=1,inplace=True)
dfe
```

```python
# now we have 25 columns(we droped 7 columns)
dfe.dtypes
```

```python
# replacing unit in mileage,power,engine
dfe['Mileage']=dfe['Mileage'].str.replace('km/kg','')  #replacing with nonspace
dfe['Mileage']=dfe['Mileage'].str.replace('kmpl','')
dfe['Engine']=dfe['Engine'].str.replace('CC','')
dfe['Power']=dfe['Power'].str.replace('bhp','')
# in engine power and mileage there are some null value(null value typed)
# replacing null
dfe['Mileage']=dfe['Mileage'].str.replace('null','0')  #we replacing with '0' because that column is object only
dfe['Engine']=dfe['Engine'].str.replace('null','0')  # at the next step we will convert it into datatype
dfe['Power']=dfe['Power'].str.replace('null','0')
dfe
```

```python
dfe.dtypes
# still datatype is not converted
```

```python
# converting datatype into float
# data type conversion
dfe['Mileage']=dfe['Mileage'].astype(float)
dfe['Engine']=dfe['Engine'].astype(float)
dfe['Power']=dfe['Power'].astype(float)
```

```python
dfe.dtypes
```

```python
# after converting datatype '0' converted to 0(but the power or mileage or engine will not be 0 )
# replacing 0 (treating as a missing value)
# replace a numerical value  (using numpy)

dfe.loc[dfe.Engine==0,'Engine']=np.NaN  #treating as a missing value(not a number(NaN))
dfe.loc[dfe.Mileage==0,'Mileage']=np.NaN
dfe.loc[dfe.Power==0,'Power']=np.NaN
```

```python
# calculating total no.of missing values
dfe.isna().sum()
# missing value has increased because there is 0
```

```python
# filling missing value
dfe['Mileage']=dfe['Mileage'].fillna(dfe['Mileage'].mean())
dfe['Engine']=dfe['Engine'].fillna(dfe['Engine'].mean())
dfe['Power']=dfe['Power'].fillna(dfe['Power'].mean())
dfe['Seats']=dfe['Seats'].fillna(dfe['Seats'].mode()[0])  #seat is filled by mode (its complete integer)
```
