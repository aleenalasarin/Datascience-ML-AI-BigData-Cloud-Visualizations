```python
 import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/melon.csv')
df
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
plt.scatter(x,y)
plt.xlabel("Weight")
plt.ylabel("Price")
plt.title("Weight-Price Graph")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_test,y_pred,color='red')
```

```python
print(model.predict([[6]]))
```

```python
print("slope is",model.coef_)
```

```python
print("constant is",model.intercept_)
```

```python
df1=pd.DataFrame({'Actual_value':y_test,'predicted_value':y_pred,'Difference':y_test-y_pred})
df1
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error
print("error is",mean_absolute_error(y_test,y_pred))
print("percentage error is",mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.metrics import r2_score
print("R2_score is",r2_score(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error
data=mean_squared_error(y_test,y_pred)
print(data)
```

```python
print(np.sqrt(data))
```
