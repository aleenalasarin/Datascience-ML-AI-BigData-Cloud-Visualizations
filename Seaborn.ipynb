```python
# seaborn is used for visualization
# in seaborn there are data set available for study purpose
# inside seaborn library a dataset is called tips
import seaborn as sns
df=sns.load_dataset('tips') #tips file load(from seaborn library)
df
```

```python
# corelation

# we can only calculate corelation between numerical values
# corelation higher value-1
# lower value-majority 0,rare case negative(-)
# we can identify which column is important or not important by looking corelation
# how one column is related to another column  =====>corelation
# corelation between same column====>1

df.corr()  #generates numerical columns
```

```python
# here totalbill efeects tip mostly  ====>(when total bill inc tip inc)
# we can drop the not important features  (while working in large file)
```

```python
# Boxplot
# lowest value at bottom
# 25percentile,50percentile,75percentile
# highest value at top
# the circles above the top highest value is called outliers
# wrong datas is called outliers (technical name is outliers)
# if the given data has wrong format then it is called outliers
# by drawing Boxplot we can find the outliers
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
# percentile

# q1:25 percentile    q2:50 percentile    q3:75 percentile
# outliers
```

```python
# 3,5,7,1,1,8,4,6

# to find 25,50,75 percentile

# first sort the data in ascending order
# 1,1,3,4,5,6,7,8

# calculate the median of the sorted data
# if it is even value make it to two parts

# part1       part2
# 1,1,3,4     5,6,7,8
# median=4+5/2=4.5

# median is q2:50 percentile

# q1 is calculated as:
# calculate the median of part1
# 1,1,3,4
# q1=3+1/2=2
# q1:25 percentile

# q3 is calculated as
# calculate the median of part2
# 5,6,7,8
# q3=6+7/2=6.5
# q3:75 percentile

# to find outliers first we calculate IQR(inter quartile range)
# equation
# IQR=q3-q1
# 6.5-2=4.5//
# 4.5 is IQR

# based on IQR we find outliers
# outliers occurs at most highest(top) and at most lowest(bootom)

# equation for top
# (1.5*IQR)+q3
# (1.5*4.5)+6.5 = 13.25//
# above 13.25 value are outliers

# equatiom for bottom
# q1-(1.5*IQR)
# 2-(1.5*4.5) = -4.75//
# if a value is below -4.75 that value will be ouliers

```
