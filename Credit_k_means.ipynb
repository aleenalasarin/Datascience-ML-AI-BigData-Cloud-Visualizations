```python
df.drop(['CUST_ID'],axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
x=df['CREDIT_LIMIT'].mean()
df['CREDIT_LIMIT'].fillna(x,inplace=True)
```

```python
y=df['MINIMUM_PAYMENTS'].mean()
df['MINIMUM_PAYMENTS'].fillna(y,inplace=True)
```

```python
df.isna().sum()
```

```python
x=df
x
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  data=KMeans(n_clusters=i,init='k-means++',random_state=42)
  data.fit(x)
  wcss.append(data.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),wcss)
plt.xlabel("No.of clusters")
plt.ylabel("Variance")
plt.title("Elbow Point")
```

```python
from sklearn.cluster import KMeans
data1=KMeans(n_clusters=6,init='k-means++',random_state=42)
data1.fit(x)
```

```python
y=data1.predict(x)
y
```

```python
x['cluster']=y
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=7)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
score
```
